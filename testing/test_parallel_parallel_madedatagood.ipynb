```python
import os
from xent.config import data_dir
from xent.datasets import SynthProcessor
from xent import M

base_dir = "parallel_parallel"

parallel_dir = os.path.join(data_dir, base_dir)
folders = [f for f in os.listdir(parallel_dir) if os.path.isdir(os.path.join(parallel_dir, f))]

processors = {
    key: SynthProcessor(
        base_dir, key
    )
    for key in folders
}

model = M("gpt2", "M0", base="base")
```

```python
for processor in processors: 
    print(f"########## Showing example from {processor}\n")
    print(model.detokenize(processors[processor].dataset[0][:300]))
    print("--------------------------- NEXT ---------------------\n\n")
```
